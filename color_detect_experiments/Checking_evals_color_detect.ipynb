```python
import altair as alt


def all_trains_in_amount(df, amount,selection):
    
    sorted_list = ['4_ids']
    chart_df_for_num_images= df[df['num_images_per_id']==amount]
    
    return alt.Chart(chart_df_for_num_images).mark_bar().encode(
    x=alt.X('run_str', sort=['4_ids', '8_ids_monocolor', '8_ids_double_colors',
       '16_ids_monocolor_and_some_permutation',
                             
       
       '16_ids_four_times_colors',
       '16_ids_four_times_permutated_colors',
        '24_ids_monocolor_and_permutated_ids',
       '24_ids_monocolor_and_non_permutated_ids',
        '32_ids_non_permutated_ids',
       '32_ids_permutated_ids',
        '64_ids_batch2_sample_num', '64_ids',]),
    
    y=alt.Y('Percent', scale=alt.Scale(domain=(0,1.0))),
    color=alt.condition(selection, 'num_images_per_id', alt.value('grey')),
    opacity=alt.condition(selection, alt.value(0.8), alt.value(0.1)),
    )

# Load the data
melted_df

# Create a selection
year_slider = alt.binding_select(options=['2','4','8','16','24','32','64','max'])
selection = alt.selection_point(fields=['num_images_per_id'],bind=year_slider)


df = melted_df[melted_df['evaluation_metric']=='percent_correct_id']
amount_imgs_per_ids = melted_df['num_images_per_id'].unique()
print(amount_imgs_per_ids)
charts = [] 

##create colors for 
for amount in amount_imgs_per_ids:
    
    charts.append(all_trains_in_amount(df, amount, selection))


# bottom = alt.Chart(df).mark_bar().encode(
# color='num_images_per_id',
#     x='num_images_per_id',

# ).add_params(selection)

layered_chart = alt.layer(*charts).add_selection(selection)


layered_chart

# chart1 = all_trains_in_amount(df, '64')
# chart1
```

```python
# param_3_num_images_per_id
# 2
import altair as alt

df = pd.read_csv('./pytorch_auto_eval_knn_reference_comparison.csv')

# Melt the DataFrame to convert it to a tidy format

melted_df = df.melt(id_vars=['run_str','num_reference_ids','num_reference_imgs_per_id','total_reference_images','reference_background','num_query_ids','total_query_imgs','query_background','model_str','num_ids_to_train_embeddor','num_imgs_per_id_to_train_embeddor','model_path','reference_file','query_file','training_file','datetime'], value_vars=['1NN', '3NN'], var_name='KNN_type', value_name='Accuracy')


# Change num_reference_imgs_per_id to *2 if reference is mixed
def change_num_ref(row):
    if row['reference_background'] == 'mix':
        row['num_reference_imgs_per_id']*=2
    return row
    
melted_df = melted_df.apply(change_num_ref, axis=1)



##change name from runstr, to be able to categorize

# def change_run_str(row):
#     string_without_sample = row['run_str'].split('_')
#     string_without_sample.pop()
#     joined= '_'.join(string_without_sample)
#     row['run_str'] = joined.replace('summer_bee_dataset_open_train_bee_','').replace('_batch1_sample_num','')
#     return row


# def change_0X_to_X(row):
#     if row['num_images_per_id'][0]=='0':
#         row['num_images_per_id']= row['num_images_per_id'][1]
        
#     return row

# melted_df = melted_df.apply(change_run_str, axis=1)
# melted_df = melted_df.apply(change_0X_to_X, axis=1)



melted_df = melted_df.drop_duplicates()
def return_knns_charts(subset, type_of_split):
    charts =[]
    
    different_types = subset['reference_background'].unique()
    
    # Create a chart with overlaid bars
    
    for reference_type in different_types:
         
        
        subset_for_reference = subset[subset['reference_background']==reference_type]
        query_count = subset_for_reference['total_query_imgs'].unique()[0]
        
            
        bar1 = alt.Chart(subset_for_reference[subset_for_reference['KNN_type'] == '1NN']).mark_bar(opacity=0.7).encode(
            x='num_reference_imgs_per_id',
            y='Accuracy',
            color=alt.value('blue')
        )
        bar2 = alt.Chart(subset_for_reference[subset_for_reference['KNN_type'] == '3NN']).mark_bar(opacity=0.7).encode(
            x='num_reference_imgs_per_id',
            y='Accuracy',
            color=alt.value('red')
        )
        chart = alt.layer(bar1, bar2).properties(
        title=reference_type+ ' query count '+str(query_count)
        )
        
        charts.append(chart)
        
    
    
    
    
    concatenated =alt.hconcat(*charts)
    
    return concatenated.properties(
    title={
        'text': type_of_split,
        'align': 'center'
    }
    )


# Load tidy to the usable DataFrame
df =melted_df[melted_df].sort_values(by='model_str')


model_trained_column = 'model_str'

# Get the unique  models
unique_models = df[model_trained_column].unique()

# Create a list of charts, one for each unique value in the column
charts = []

# Knn different knn's

knns= ['1NN', '3NN']

for model in unique_models:
    
    
    #create a chart per
    
    subset=df[df[model_trained_column] == model]
    
    display(return_knns_charts(subset, model))
    
    
    
```
